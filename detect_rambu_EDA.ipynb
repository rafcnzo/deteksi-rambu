```python
import os
import shutil

# Direktori tujuan
dest_train_img = "images/train/"
dest_train_label = "images/labels/train/"

# Membuat direktori tujuan jika belum ada
os.makedirs(dest_train_img, exist_ok=True)
os.makedirs(dest_train_label, exist_ok=True)

# Menyalin file
for f, f_label in zip(f_train_img, f_train_label):
    print(f"Menyalin {f} ke {dest_train_img}")
    shutil.copy(f, dest_train_img)
    shutil.copy(f_label, dest_train_label)
```

```python
import os
import shutil

# Direktori tujuan
dest_test_img = "images/test/"
dest_test_label = "images/labels/test/"

# Membuat direktori tujuan jika belum ada
os.makedirs(dest_test_img, exist_ok=True)
os.makedirs(dest_test_label, exist_ok=True)

# Menyalin file
for f, f_label in zip(f_test_img, f_test_label):
    dest_img_path = os.path.join(dest_test_img, os.path.basename(f))  # Path lengkap untuk file gambar
    dest_label_path = os.path.join(dest_test_label, os.path.basename(f_label))  # Path lengkap untuk file label
    
    print(f"Menyalin {f} ke {dest_img_path}")
    shutil.copy(f, dest_img_path)
    shutil.copy(f_label, dest_label_path)
```

```python
import pandas as pd

df = pd.DataFrame(columns = ['class', 'label'] )
for f in f_train_label:
    f_base = os.path.basename(f)
    class_name = f_base.split(" (",1)[0]
    df_label = pd.read_csv(f,sep=' ', header=None, names=['class','x', 'y', 'w', 'h'])
    df_label['label'] = class_name
    df = pd.concat([pd.DataFrame(df_label[['class','label']]), df], ignore_index=True)

#print(df.drop_duplicates()['class'][:1])
df = df.drop_duplicates()
df = df.sort_values(by=['class'])

delimiter=':'
for row in df.itertuples(index=False, name=None):
    print(f"{row[0]}: {row[1]}")
```

```python
import shutil
import os
from glob import glob

# Folder tujuan
dest_val_img = "images/val/"
dest_val_label = "labels/val/"

# Membuat folder tujuan jika belum ada
os.makedirs(dest_val_img, exist_ok=True)
os.makedirs(dest_val_label, exist_ok=True)

# File sumber dengan pola
f_val_img = glob("images/test/*[78]*.jpg")
f_val_label = glob("labels/test/*[78]*.txt")

# Memindahkan file gambar
for f in f_val_img:
    print(f"Memindahkan gambar: {f}")
    shutil.move(f, os.path.join(dest_val_img, os.path.basename(f)))

# Memindahkan file label
for f in f_val_label:
    print(f"Memindahkan label: {f}")
    shutil.move(f, os.path.join(dest_val_label, os.path.basename(f)))

print("Semua file telah berhasil dipindahkan.")
```
